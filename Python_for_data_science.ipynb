**Merging** The merge function alows you to merge dataframe together using a similar logic as merging SQL tables together

```python
import pandas as pd

# Assuming 'left' DataFrame is defined like this:
left = pd.DataFrame({'Key': ['K0', 'K1', 'K2', 'K3'],
                   'A': ['A0', 'A1', 'A2', 'A3'],
                   'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'Key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})

# Now, the merge operation should work:
pd.merge(left, right, how='inner', on='Key')
```

**Joining**

```python
import pandas as pd

# Creating the 'left' DataFrame from a dictionary
left_data = {
    'Key': ['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
}
left = pd.DataFrame(left_data)

# Creating the 'right' DataFrame from a dictionary
right_data = {
    'Key': ['K1', 'K2', 'K3', 'K4'],
    'C': ['C1', 'C2', 'C3', 'C4'],
    'D': ['D1', 'D2', 'D3', 'D4']
}
right = pd.DataFrame(right_data)
```

```python
outer_join = left.join(right.set_index('Key'), on='Key', how='outer')
print("Outer Join:\n", outer_join)
```

```python
left_join = left.join(right.set_index('Key'), on='Key', how='left')
print("Left Join:\n", left_join)
```

```python
right_join = left.join(right.set_index('Key'), on='Key', how='right')
print("Right Join:\n", right_join)
```

```python
inner_join = left.join(right.set_index('Key'), on='Key', how='inner')
print("Inner Join:\n", inner_join)
```
